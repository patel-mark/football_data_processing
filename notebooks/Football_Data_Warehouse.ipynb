```python
# %% [markdown]
### Football Data Warehouse Pipeline - Jupyter Notebook Version
# %%
import pandas as pd
import numpy as np
import glob
import os
from pathlib import Path
import re

# For visualization (optional)
import matplotlib.pyplot as plt
import seaborn as sns

# Set up paths
BASE_DIR = Path.cwd()
DATA_DIR = Path("..") / "data"
#DATA_DIR = BASE_DIR / "data"
OUTPUT_DIR = BASE_DIR / "output"
OUTPUT_DIR.mkdir(exist_ok=True)
```

```python
csv_dataframes = {}

# Loop through each league folder
for league_folder in DATA_DIR.iterdir():
    if league_folder.is_dir():
        for csv_file in league_folder.glob("*.csv"):
            # Use filename (without .csv) as key
            df_key = f"{league_folder.name}/{csv_file.stem}"
            csv_dataframes[df_key] = pd.read_csv(csv_file)
```
